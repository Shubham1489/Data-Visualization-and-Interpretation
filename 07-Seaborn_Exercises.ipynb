```python
# Import required libraries
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Load built-in datasets
tips = sns.load_dataset("tips")
flights = sns.load_dataset("flights")
iris = sns.load_dataset("iris")

# Preview datasets
print(tips.head())
print(flights.head())
print(iris.head())
```

```python
# Histogram of total_bill
sns.histplot(tips['total_bill'], kde=True, bins=20)
plt.title("Distribution of Total Bill (Tips Dataset)")
plt.show()

# KDE Plot for iris petal_length
sns.kdeplot(iris['petal_length'], fill=True)
plt.title("KDE Plot of Petal Length (Iris Dataset)")
plt.show()
```

```python
# Scatter plot of total_bill vs tip
sns.scatterplot(x="total_bill", y="tip", data=tips, hue="sex")
plt.title("Total Bill vs Tip (Tips Dataset)")
plt.show()

# Iris dataset scatter plot
sns.scatterplot(x="sepal_length", y="sepal_width", data=iris, hue="species")
plt.title("Sepal Length vs Width by Species (Iris Dataset)")
plt.show()
```

```python
sns.barplot(x="day", y="total_bill", data=tips, errorbar=None)
plt.title("Average Total Bill by Day (Tips Dataset)")
plt.show()
```

```python
sns.histplot(tips['tip'], kde=True, bins=15)
plt.title("Distribution of Tips")
plt.show()

sns.kdeplot(tips['total_bill'], fill=True, color="green")
plt.title("KDE Plot of Total Bill")
plt.show()
```

```python
# Barplot
sns.barplot(x="sex", y="tip", data=tips)
plt.title("Average Tip by Gender")
plt.show()

# Countplot
sns.countplot(x="day", data=tips, hue="sex")
plt.title("Count of Customers by Day and Gender")
plt.show()

# Boxplot
sns.boxplot(x="day", y="total_bill", data=tips)
plt.title("Boxplot of Total Bill by Day")
plt.show()

# Violinplot
sns.violinplot(x="day", y="tip", data=tips, hue="sex", split=True)
plt.title("Violin Plot of Tips by Day and Gender")
plt.show()

# Swarmplot
sns.swarmplot(x="day", y="tip", data=tips)
plt.title("Swarm Plot of Tips by Day")
plt.show()
```

```python
sns.regplot(x="total_bill", y="tip", data=tips)
plt.title("Regression Plot: Total Bill vs Tip")
plt.show()

sns.lmplot(x="total_bill", y="tip", data=tips, hue="sex", col="time")
plt.suptitle("LMPlot: Tips by Gender and Time", y=1.05)
plt.show()
```

```python
# PairGrid on iris
g = sns.PairGrid(iris, hue="species")
g.map_diag(sns.histplot)
g.map_offdiag(sns.scatterplot)
g.add_legend()
plt.show()

# FacetGrid on tips
g = sns.FacetGrid(tips, col="sex", row="time", hue="smoker")
g.map(sns.scatterplot, "total_bill", "tip")
g.add_legend()
plt.show()

# JointGrid
g = sns.JointGrid(data=tips, x="total_bill", y="tip")
g.plot(sns.scatterplot, sns.histplot)
plt.show()
```

```python
sns.set_style("darkgrid")
sns.scatterplot(x="total_bill", y="tip", data=tips)
plt.title("Scatter Plot with Darkgrid Style")
plt.show()

sns.set_style("whitegrid")
sns.boxplot(x="day", y="tip", data=tips)
plt.title("Boxplot with Whitegrid Style")
plt.show()

sns.set_style("ticks")
sns.violinplot(x="sex", y="tip", data=tips)
plt.title("Violin Plot with Ticks Style")
plt.show()
```

```python
sns.set_palette("husl")
sns.countplot(x="day", data=tips)
plt.title("Countplot with Husl Palette")
plt.show()

sns.set_palette("coolwarm")
sns.barplot(x="sex", y="tip", data=tips)
plt.title("Barplot with Coolwarm Palette")
plt.show()
```
